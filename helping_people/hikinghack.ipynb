```python
#helping out @hikinghack
#see https://twitter.com/HikingHack/status/1481353359294701570

import numpy as np
np.random.seed(42)
```

```python
#create two dummy datasets of length N
N = 30

#create the two datasets a dnd b - each datasets has 1000 rows and 2 columns with random numbers between 1 and 10
a = np.random.randint(40, 44, size=(N, 2))
b = np.random.randint(40, 44, size=(N, 2))
ix = np.array(np.arange(N))
```

```python
#add two columns to the our dataset. The first column is the index of the row (From 0 to 199) and the second column is 0 for dataset a and 1 for dataset b
a = np.c_[ix, np.zeros(N), a]
b = np.c_[ix, np.ones(N), b]
```

```python
#now add the two datasets together
c = np.r_[a,b]
```

```python
#sort by the last two column
c = c[np.lexsort((c[:,3], c[:,2]))]
c
```

```python
#creates a sibling array shifted by one row up
c1 = np.roll(c,-1, axis=0)
c1
```

```python
#subtract the two
d = c1-c
#then take only the last two columns 
e = d[:,2:4] 
e
```

```python
#now sum the columns row by row
f = np.sum(e, axis=1)

#find the indeces of the rows where the sum is 0
idx = np.where(f==0)
```

```python
#now show the rows that match
r = c1[idx]
r
```

```python
#you may sort back by the first two columns now if that helps visualisation
r = r[np.lexsort((r[:,0], r[:,1]))]
r
```
